## Análise dos resultados

### Conceitos

1. Overfitting:

Overfitting acontece quando o modelo aprende tão bem os detalhes e o ruído do conjunto de treinamento que ele começa a ter um desempenho ruim em dados novos e não vistos (o conjunto de validação ou teste).  Deste modo, é como se o  modelo "decorasse" as respostas do treino em vez de aprender a generalizar os padrões.


2. Generalização Limitada:

A capacidade do modelo de aplicar o conhecimento aprendido em dados de treinamento para dados novos e diversos.


3. Treinamento Instável:

O processo de treinamento pode se tornar mais instável e sensível a pequenas mudanças nos dados ou nos hiperparâmetros do treinamento.

### Análise


A análise dos resultados de validação revela um mAP50 geral de 93.1% e um mAP50-95 de 72.8%, indicando alta precisão nas detecções e uma boa qualidade das caixas delimitadoras. A detecção de drones se destaca com métricas de precisão e revocação elevadas, enquanto a detecção de tratores apresenta uma ligeira menor revocação.

A análise comparativa das precisões ao longo de 120 épocas de treinamento mostra uma estabilidade na precisão geral (mAP50) após um pico inicial. A precisão para drones se mantém consistentemente alta. A detecção de tratores exibe maior variabilidade entre as épocas. Ao considerar a métrica mais rigorosa (mAP50-95), observa-se que a época 90 frequentemente apresenta os melhores resultados, especialmente para a detecção geral e de tratores.

Portanto, o sistema mostra potencial na detecção de tratores e drones, com alta precisão, mas o dataset pequeno causa variabilidade (principalmente em tratores) e possível overfitting (queda no mAP50-95 na validação). Drones são detectados com mais facilidade. Sendo assim,  o modelo que apresentou melhor resultado de forma geral foi o de 90 épocas por equilibrar precisão geral e localização (melhor mAP50-95 para tratores), minimizando o overfitting. A expansão do dataset é crucial para o desempenho futuro.
